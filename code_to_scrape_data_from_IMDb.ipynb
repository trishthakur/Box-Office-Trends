```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
import random
```

```python
movie_number_lst = []
movie_name_lst = []
movie_year_lst = []
movie_runtime_lst = []
movie_genre_lst = []
movie_rating_lst = []
movie_votes_lst = []
movie_revenue_lst = []
movie_country_lst = []
```

```python
while len(movie_number_lst) <= 1000:

    page_no = random.randint(1, 50)
    url = 'https://www.imdb.com/list/ls063676189/?sort=list_order,asc&st_dt=&mode=detail&page={}'.format(page_no)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    data_soup = soup.find_all('div', class_= 'lister list detail sub-list')[0].find_all('div', 
                class_ = 'lister-list')[0].find_all('div', class_ = 'lister-item mode-detail')
    movies_in_page = len(data_soup)
    movie_no = random.randint(0, movies_in_page)
    if movie_no == 100:
        movie_no = movie_no - 1
    movie_soup = data_soup[movie_no]

    # get movie number
    movie_no = movie_soup.find('span', class_ = 'lister-item-index unbold text-primary').get_text().split('.')[0]
    
    if movie_no not in movie_number_lst:
        movie_number_lst.append(movie_no)
        
        # get movie tconst value
        movie_tconst = movie_soup.find('div', class_ = 'lister-item-image')['data-tconst']

        # get movie name
        movie_name = movie_soup.find_all('a')[1].get_text()
        movie_name_lst.append(movie_name)
        
        # get year
        try:
            movie_year = movie_soup.find('span', class_ = 'lister-item-year text-muted unbold').get_text()[1:-1]
            movie_year_lst.append(movie_year)
        except:
            movie_year_lst.append('NA')
        
        # get runtime
        try:
            movie_runtime = movie_soup.find('span', class_ = 'runtime').get_text().split(' ')[0]
            movie_runtime_lst.append(movie_runtime)
        except:
            movie_runtime_lst.append('NA')
        
        # get genre
        try:
            movie_genre = movie_soup.find('span', class_ = 'genre').get_text().replace('\n', '').split(',')[0]
            movie_genre_lst.append(movie_genre)
        except:
            movie_genre_lst.append('NA')
        
        # get rating
        try:
            movie_rating = movie_soup.find('span', class_ = 'ipl-rating-star__rating').get_text()
            movie_rating_lst.append(movie_rating)
        except:
            movie_rating_lst.append('NA')
        
        # get votes
        try:
            movie_votes = movie_soup.find_all('p', class_ = 'text-muted text-small')[2].find_all('span', 
                                    {'data-value':True})[0]['data-value']
            movie_votes_lst.append(movie_votes)
        except:
            movie_votes_lst.append('NA')
        
        # get gross revenue
        try:
            movie_revenue = movie_soup.find_all('p', class_ = 'text-muted text-small')[2].find_all('span', 
                                    {'data-value':True})[1]['data-value']
            movie_revenue_lst.append(movie_revenue)
        except:
            movie_revenue_lst.append('NA')
        
        
        url2 = 'https://www.imdb.com/title/{}/?ref_=ttls_li_tt'.format(movie_tconst)
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
                  }
        response = requests.get(url2, headers = headers)
        soup = BeautifulSoup(response.text, "html.parser")

        # get country of origin
        try:
            movie_country = soup.find_all('section', class_ = 
                'ipc-page-section ipc-page-section--base celwidget', attrs = {'data-testid':"Details"})[0].find_all('a', 
                class_ = 'ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link')[1].get_text()
            movie_country_lst.append(movie_country)
        except:
            movie_country_lst.append('NA')
        
        print('Movies scrapped:', len(movie_number_lst))
    
    else:
        continue
```

```python
data_dict = {
    'movie_name': movie_name_lst,
    'release_year': movie_year_lst,
    'runtime': movie_runtime_lst,
    'genre': movie_genre_lst,
    'rating': movie_rating_lst,
    'votes': movie_votes_lst,
    'revenue': movie_revenue_lst,
    'country': movie_country_lst
}
```

```python
df = pd.DataFrame(data_dict)
df
```
